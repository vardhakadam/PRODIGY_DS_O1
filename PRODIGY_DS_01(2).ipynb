### Plotting Bar Chart 

```python
years = df.columns[1:]  

total_values = df[years].sum()

plt.figure(figsize=(30, 30))  
plt.barh(years, total_values,color='#191970')
plt.xlabel('Total Values', size=30)
plt.ylabel('Year', size=30)
plt.title('Total Values per Year', size=45)
plt.show()
```
